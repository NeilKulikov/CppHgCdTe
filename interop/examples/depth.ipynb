```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import hgcdte_py as hgcdte
```

```python
def make_well(beg, width, xb, xw, delta = 1.e-5):
    zs = np.array([beg, beg + delta, beg + width - delta, beg + width])
    xs = np.array([xb,  xw,          xw,                xb])
    return (zs, xs)
```

```python
mds = []
xb = 0.7
xs = np.array([0.01 * float(x) for x in range(0, 100)])
for x in xs:
    zs = [0., 10., 10.001, 14.999, 15., 25.]
    ys = [xb, xb,  x,      x,      xb,  xb ]
    md = hgcdte.model(np.array(zs), np.array(ys))
    mds.append(md)
```

```python
len(xs), len(mds)
```

```python
hcs = [hgcdte.hcore(m) for m in mds]
```

```python
sps = np.array([h.spectre((0., 0.)) for h in hcs])
```

```python
plt.figure(figsize=(7.5,5), dpi=120)
plt.plot(xs, sps[:, 366], 'y-')
plt.plot(xs, sps[:, 368], 'r-')
plt.plot(xs, sps[:, 370], 'g-')
plt.plot(xs, sps[:, 372], 'b-')
plt.plot(xs, sps[:, 374], 'k-')
plt.grid()
```
