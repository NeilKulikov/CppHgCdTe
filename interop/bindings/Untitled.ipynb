```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as interp
```

```python
def func(x, y):
    return x*(1-x)*np.cos(4*np.pi*x) * np.sin(4*np.pi*y**2)**2
```

```python
grid_x, grid_y = np.mgrid[-1:2:500j, -1:2:500j]
```

```python
grid_f = func(grid_x, grid_y)
```

```python
plt.imshow(grid_f)
```

```python
points = np.random.rand(1000, 2)
values = func(points[:,0], points[:,1])
```

```python
plt.figure(figsize = (5, 5), dpi = 120)
plt.imshow(grid_f.T, extent=(-1,2,-1,2), origin='lower')
plt.plot(points[:,0], points[:,1], 'k.', ms=1)
```

```python
grid_z = interp.griddata(points, values, (grid_x, grid_y), method='cubic')
```

```python
plt.figure(figsize = (5, 5), dpi = 120)
plt.imshow(grid_z.T, extent=(-1,2,-1,2), origin='lower')
plt.plot(points[:,0], points[:,1], 'k.', ms=1)
```

```python
plt.figure(figsize = (5, 5), dpi = 120)
plt.imshow(grid_f - grid_z)
```
