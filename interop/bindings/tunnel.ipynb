```python
xs = np.array([0.0, 10.0, 10.0001, 14.9999, 15.0, 20.0, 20.0001, 24.9999, 25.0, 35.0])
ys = np.array([0.7, 0.7,  0.1,     0.1,     0.7,  0.7,  0.1,     0.1,     0.7,  0.7])
```

```python
hi = hgcdte.make_hinst(hc, (0.0, 0.0))
```
