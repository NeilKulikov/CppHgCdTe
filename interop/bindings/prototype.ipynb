```julia
include("./hgcdte_julia.jl")
```

```julia
using .HgCdTe
```

```julia
xs = [ 0.; 11.0; 11.000001; 13.999999; 14.0; 25. ]
ys = [ 0.7; 0.7; 0.1; 0.1; 0.7; 0.7]
```

```julia
md = Model(xs, ys)
```

```julia
hc = HCore(md, UInt64(61))
```

```julia
ks = collect(0.:0.05:1.)
```

```julia
hs = [HInst(hc, (kx, 0.)) for kx in ks]
```

```julia
es = Eigens.(hs)
```

```julia
cs4 = Band.(es, 3)
cs3 = Band.(es, 2)
cs2 = Band.(es, 1)
cs1 = Band.(es, 0)
vs1 = Band.(es, -1)
vs2 = Band.(es, -2)
vs3 = Band.(es, -3)
vs4 = Band.(es, -4)
```

```julia
using PyPlot
```

```julia
plot(ks, vs1)
plot(ks, vs2)
plot(ks, vs3)
plot(ks, vs4)
plot(ks, cs1)
plot(ks, cs2)
plot(ks, cs3)
plot(ks, cs4)
grid("on")
```
