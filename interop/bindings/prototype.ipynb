```julia
include("./hgcdte_julia.jl")
```

```julia
using .HgCdTe
```

```julia
xs = [ 0.; 9.9; 10.1; 14.9; 15.1; 25. ]
ys = [ 0.7; 0.7; 0.1; 0.1; 0.7; 0.7]
```

```julia
md = Model(xs, ys)
```

```julia
hc = HCore(md, UInt64(61))
```

```julia
ks = collect(0.:0.05:1.)
```

```julia
hs = [HInst(hc, (kx, 0.)) for kx in ks]
```

```julia
es = Eigens.(hs)
```

```julia
vs1 = [x[368] for x in es]
vs2 = [x[370] for x in es]
vs3 = [x[372] for x in es]
cs1 = [x[374] for x in es]
cs2 = [x[376] for x in es]
```

```julia
using PyPlot
```

```julia
plot(ks, vs1)
plot(ks, vs2)
plot(ks, vs3)
plot(ks, cs1)
grid("on")
```
