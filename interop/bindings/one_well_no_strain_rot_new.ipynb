```python
import math as m
```

```python
import hgcdte
```

```python
import matplotlib.pyplot as plt
```

```python
xs = [0.0, 0.0001, 3., 3.0001, 20.]
ys = [0.65, 0.1,   0.1,  0.65,   0.65]
```

```python
plt.plot(xs, ys)
plt.grid()
```

```python
rot = hgcdte.rotator(hkl = [0, 1, 3])
mod = hgcdte.model(xs, ys)
```

```python
hcr = hgcdte.hcore(mod, 41, rot = rot)
```

```python
ks2 = np.arange(0., 1.02, 0.02)
```

```python
eiss2 = np.array([hcr.spectre((kx, 0.)) for kx in ks2])
```

```python
plt.figure(figsize=(7.5,5), dpi=120)
plt.plot(ks2, eiss2[:, 248], 'k-')
plt.plot(ks2, eiss2[:, 246], 'k-')
plt.plot(ks2, eiss2[:, 244], 'y-')
plt.plot(ks2, eiss2[:, 240], 'r-')
plt.plot(ks2, eiss2[:, 238], 'g-')
plt.plot(ks2, eiss2[:, 236], 'b-')
plt.grid()
```
